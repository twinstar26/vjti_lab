1.   Create a function that multiplies any number x by 13


*   x=14
*   if result is greater than 130 print "x is greater than 10"
*   else print "x is less than 10"





```
### START CODE HERE ### 
def mul(a):
  return a*13

x = mul(14)
if x>130:
  print ("x is greater than 10")
else:
  print("x is less than 10")


### END CODE HERE ###
```
